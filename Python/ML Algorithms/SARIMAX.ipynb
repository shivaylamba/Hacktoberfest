# Just using SARIMA here not consdering X the exogenous variables till now

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
len(df1)
```

```python
train = df1.iloc[:436]
test = df1.iloc[436:]
```

```python
from pmdarima import auto_arima
```

```python
auto_arima(df1['total'],seasonal=True,m=7,trace=True).summary()
```

```python
# Value Error : non-invertible starting MA parameters found
model = SARIMAX(train['total'],order = (0,0,0),seasonal_order=(1,0,1,7),enforce_invertibility=False)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
start = len(train)
end = len(train) + len(test) -1
```

```python
predictions = results.predict(start,end).rename('SARIMA Model')
```

```python
ax = test['total'].plot(legend=True,figsize = (15,8))
predictions.plot(legend=True)

for day in df1[df1['holiday']==1].index:
    ax.axvline(x=day,color='g',alpha=0.9,linestyle='--');

```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
error = rmse(test['total'],predictions)
```

```python
error
```

# TRAIN ON SARIMAX MODEL

```python
model = SARIMAX(train['total'],exog=train[['holiday']],order=(1,0,0),seasonal_order=(1,0,1,7),enforce_invertibility=False)
# exog accepts a dataframe as like above in X
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
start = len(train)
end = len(train)+len(test)-1
```

```python
predictions = result.predict(start,end,exog = test[['holiday']]).rename('SARIMAX with Exogenous')
```

```python
predictions.plot(figsize = (12,8),legend=True)
test['total'].plot(legend=True);
```

```python
ax = test['total'].plot(legend=True,figsize = (15,8))
predictions.plot(legend=True)

for day in df1[df1['holiday']==1].index:
    ax.axvline(x=day,color='g',alpha=0.9,linestyle='--');

```

```python
rmse(test['total'],predictions)
```

<i><font color ='red'>***as u can see the error is decreased so this model is better but now the model performs not very well on non-holidays but still its a better model as rmse is reduced***</i>
